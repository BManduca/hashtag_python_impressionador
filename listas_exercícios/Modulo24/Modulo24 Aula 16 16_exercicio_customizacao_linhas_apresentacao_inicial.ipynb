```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

# Palheta de cores Set3
cores = plt.get_cmap('Set3').colors
# parametros cycler -> 'color' para alterar a cor e a variavel criada contendo a paleta de cores
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

cor_anotacoes = cores[4]

# Tamanho 14x6
plt.figure(figsize=(14,6))

# Linha com espessura 3
plt.plot(anos, refeicoes, linewidth=3, zorder=1)

# Destaque os pontos de início e fim da série com um scatter plot com as quantidades de refeições servidas
for posicao in (0, -1):
    plt.scatter(anos[posicao], refeicoes[posicao], color=cores[3], zorder=2)
    plt.annotate(refeicoes[posicao], color=cor_anotacoes, xy=(anos[posicao], refeicoes[posicao]), xytext=(5, 0), textcoords='offset points', size=12, weight='bold')

for posicao in (10, 11):
    plt.scatter(anos[posicao], refeicoes[posicao], color=cores[3], zorder=2)
    plt.annotate(refeicoes[posicao], color=cor_anotacoes, xy=(anos[posicao], refeicoes[posicao]), xytext=(14, 0), textcoords='offset points', size=11, weight='bold')

# Destaque o impacto da pandemia (2020 e 2021) com um vspan
plt.axvspan(2020, 2021, color=cores[5], alpha=0.5, zorder=0)

# Destaque o impacto da pandemia com um texto
plt.text(2020, 40000, 'Pandemia', fontsize=11, color=cor_anotacoes, weight='bold')


# Remova os eixos y
# Remova os ticks do eixo x
# Remova as bordas do gráfico
plt.yticks([])
# mostrando todos os anos presentes no eixo x
plt.xticks(anos)
plt.tick_params(axis='x', length=0)
plt.box(False)

# Adicione um título
plt.title('Refeições servidas por ano', weight='bold', fontsize=16, loc='left', color='gray')

# A figura deve ser salva numa pasta chamada `imagens` que deve estar no mesmo diretório do Jupyter Notebook
plt.savefig('imagens/grafico_linhas_refeicoes_pandemia_ano.png', bbox_inches='tight', dpi=300)

plt.show()
```
