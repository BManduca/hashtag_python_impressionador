#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

````python
nome = input('Nome: ')

info = []

while len(nome) <= 3:
    print('Nome incorreto. Favor inserir um nome com mais de 3 caracteres.')
    nome = input('Nome: ')
info.append(nome)

idade = int(input('Informe a sua idade(0 a 150): '))
while not (0 <= idade <= 150):
    print('Favor inserir a idade entre 0 e 150 anos!')
    idade = int(input('Informe a sua idade(0 a 150): '))
info.append(idade)

salario = float(input('Informe o salario (obs.: o valor precisa ser maior que 0): '))
while salario <= 0:
    print('Favor inserir um valor de salário maior que zero.')
    salario = float(input('Informe o salario (obs.: o valor precisa ser maior que 0): '))
info.append(salario)

sexo = input('Informe o sexo (m - masculino / f - feminino): ').lower()
while sexo not in ['m', 'f']:
    print('Favor inserir valor correto para o sexo/gênero.')
    sexo = input('Informe o sexo (m - masculino / f - feminino): ').lower()
info.append(sexo)

estado_civil = input('Informe o estado civil ["s", "c", "v", "d"]: ').lower()
while estado_civil not in ['s', 'c', 'v', 'd']:
    print('Favor inserir um valor correto para o estado civil!')
    estado_civil = input('Informe o estado civil ["s", "c", "v", "d"]: ').lower()
info.append(estado_civil)

print(f'\n{info}')
````

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

````python
faturamentos = []

text = 'ANÁLISE DOS ÚLTIMOS 5 MESES'

print(f'{text:^80}\n')

for i in range(5):
    faturamento = float(input(f'Informe o valor do faturamento do mês [{i+1}]: '))
    faturamentos.append(faturamento)

faturamento_total = sum(faturamentos)

faturamento_medio = faturamento_total / len(faturamentos)

print(f'Faturamentos dos últimos 5 meses: {faturamentos}\n')

print(f'Faturamento total: R$ {faturamento_total:,}\n')

print(f'Faturamento médio por mês: R$ {faturamento_medio:,.2f}')

````

#### 12. Um funcionário de uma empresa recebe aumento salarial anualmente: Sabe-se que:
<pre>
Esse funcionário foi contratado em 1995, com salário inicial de R$ 1.000,00;
Em 1996 recebeu aumento de 1,5% sobre seu salário inicial;
A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior.
Faça um programa que determine o salário desse funcionário em 2003. 
</pre>

````python
salario = 1000 * 1.015
porcentagem_aumento = 0.015

for i in range(1997, 2004):
    porcentagem_aumento *= 2
    salario *= (porcentagem_aumento + 1)

print(f'O Salario do funcionário em 2003 foi de R$ {salario:,.2f}')
````

Bauru com ovo   102     R$ 1,50
Hambúrguer      103     R$ 1,20

Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido.
Considere que o cliente deve informar quando o pedido deve ser encerrado.
</pre>

````python

total = 0

codigo_produto = input('Informe o código do produto, caso deseje encerra, basta digitar finalizar): ')

while codigo_produto != 'finalizar':
    qtde_produto = int(input(f'Informe a quantidade do produto {codigo_produto}: '))

    if codigo_produto == '100':
        total += qtde_produto * 1.20
    elif codigo_produto == '101':
        total += qtde_produto * 1.30
    elif codigo_produto == '102':
        total += qtde_produto * 1.50
    elif codigo_produto == '103':
        total += qtde_produto * 1.20
    elif codigo_produto == '104':
        total += qtde_produto * 1.30
    elif codigo_produto == '105':
        total += qtde_produto * 1.00

    codigo_produto = input('Informe o código do produto, caso deseje encerra, basta digitar finalizar): ')


print(f'Total do pedido foi de R$ {total:,.2f}')
````

#### 14. Uma grande emissora de televisão quer fazer uma enquete entre os seus telespectadores para saber qual o melhor jogador após cada jogo. Para isto, faz-se necessário o desenvolvimento de um programa, que será utilizado pelas telefonistas, para a computação dos votos. Sua equipe foi contratada para desenvolver este programa, utilizando a linguagem de programação Python. Para computar cada voto, a telefonista digitará um número, entre 1 e 23, correspondente ao número da camisa do jogador. Um número de jogador igual zero, indica que a votação foi encerrada. Se um número inválido for digitado, o programa deve ignorá-lo, mostrando uma breve mensagem de aviso, e voltando a pedir outro número. Após o final da votação, o programa deverá exibir:
<pre>a. O total de votos computados;
b. Os números e respectivos votos de todos os jogadores que receberam votos;
c. O percentual de votos de cada um destes jogadores;
d. O número do jogador escolhido como o melhor jogador da partida, juntamente com o número de 
votos e o percentual de votos dados a ele.
        Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece
        ordenado pelo número do jogador. O programa deve fazer uso de arrays. O programa deverá executar
        o cálculo do percentual de cada jogador através de uma função. Esta função receberá dois parâmetros:
        o número de votos de um jogador e o total de votos. A função calculará o percentual e retornará
        o valor calculado. Abaixo segue uma tela de exemplo. O disposição das informações deve ser o mais
        próxima possível ao exemplo. Os dados são fictícios e podem mudar a cada execução do programa. Ao final,
        o programa deve ainda gravar os dados referentes ao resultado da votação em um arquivo texto no disco,
        obedecendo a mesma disposição apresentada na tela.</pre>
```
Enquete: Quem foi o melhor jogador?

Número do jogador (0=fim): 9
Número do jogador (0=fim): 10
Número do jogador (0=fim): 9
Número do jogador (0=fim): 10
Número do jogador (0=fim): 11
Número do jogador (0=fim): 10
Número do jogador (0=fim): 50
Informe um valor entre 1 e 23 ou 0 para sair!
Número do jogador (0=fim): 9
Número do jogador (0=fim): 9
Número do jogador (0=fim): 0

Resultado da votação:

Foram computados 8 votos.

Jogador Votos           %
9               4               50,0%
10              3               37,5%
11              1               12,5%
``` 

````python
votos = []
print('Enquete: Quem foi o melhor jogador?')

for i in range(23):
    votos.append(0)

while True:
    num_jogador = int(input('Informe o número do jogador(0 = Fim): '))
    if num_jogador == 0:
        break
    elif not 1 <= num_jogador <= 23:
        print('Número inválido.\nInforme o número do jogador entre 1 e 23 ou 0 para finalizar.')
        continue
    votos[num_jogador - 1] += 1

total_votos = sum(votos)

print(f'\nForam computados {total_votos} votos.\n')
print('Jogador\tVotos\t%')
for i, voto in enumerate(votos):
    print(f'{i+1}\t{voto}\t{voto/total_votos:.1%}')

votos_melhor_jogador = max(votos)
ind_melhor_jogador = votos.index(votos_melhor_jogador)
porcentagem = votos_melhor_jogador / total_votos

print(f'\n\nO melhor jogador foi o {ind_melhor_jogador+1}, com {votos_melhor_jogador} votos, que corresponde a {porcentagem:.1%} do total de votos')
````
