#### 1. Faça um Programa que leia as vendas dos vendedores, mostre a venda de cada vendedor com o seu nome e a média de vendas. 

````python
vendas = [1000, 1500, 1200, 1300]
vendedores = ["Fulano", "Beltrano", "Ciclano", "Lira"]

for i in range(len(vendas)):
    print(f'Vendedor: {vendedores[i]} | Vendas: {vendas[i]}')

media = sum(vendas) / len(vendas)

print(f'\nA média de vendas foi de {media}')
````

#### 2. Faça um Programa que crie uma lista com as médias de cada aluno, imprima as médias de cada aluno e a quantidade de alunos com média maior que 7.

````python
alunos = ["José", "Joana", "Maria", "Carla", "Mauricio", "Andre", "Tiago", "Enzo", "Amanda", "Alessandra"]

medias_alunos = []

notas = [
    [10, 9, 8, 8],
    [9, 7, 6, 4],
    [10, 10, 10, 10],
    [5, 3, 10, 9],
    [7, 6, 6, 6],
    [7, 7, 8, 7],
    [7, 7, 7, 9],
    [8, 5, 6, 7],
    [10, 9, 7, 4],
    [10, 1, 3, 3],
]

for i in range(len(notas)):
    media = sum(notas[i]) / len(notas[i])
    medias_alunos.append(media)

contador_medias = 0

for media in medias_alunos:
    if media > 7:
        contador_medias += 1

for i, media in enumerate(medias_alunos):
    print(f'Aluno: {alunos[i]} | Média: {media}')

print(f'\n\n{contador_medias} alunos tiveram média maior que 7')
````

#### 3. Foram anotadas as idades e salários de 30 funcionários. Faça um programa que determine quantos funcionários com mais de 25 anos possuem salário inferior à média de todos os salários.

#### 5. As Organizações Tabajara resolveram dar um abono aos seus colaboradores em reconhecimento ao bom resultado alcançado durante o ano que passou. Para isto contratou você para desenvolver a aplicação que servirá como uma projeção de quanto será gasto com o pagamento deste abono.
<pre>Após reuniões envolvendo a diretoria executiva, a diretoria financeira e os representantes do sindicato laboral, <br>chegou-se a seguinte forma de cálculo:
. Cada funcionário receberá o equivalente a 20% do seu salário bruto de dezembro;
. O piso do abono será de 100 reais, isto é, aqueles funcionários cujo salário for muito baixo, recebem <br>este valor mínimo; 

Neste momento, não se deve ter nenhuma preocupação com colaboradores com tempo menor de casa, descontos, <br>impostos ou outras particularidades. O programa <br>deverá calcular o valor do abono concedido a cada colaborador, de acordo com a regra definida acima. <br>Ao final, o programa deverá apresentar:</pre>
<pre>
O salário de cada funcionário, juntamente com o valor do abono;
O número total de funcionário processados;
O valor total a ser gasto com o pagamento do abono;
O número de funcionário que receberá o valor mínimo de 100 reais;
O maior valor pago como abono; 
<br>
A tela abaixo é um exemplo de execução do programa, apenas para fins ilustrativos.
<pre>
Projeção de Gastos com Abono
============================ 
 
Salário: 1000
Salário: 300
Salário: 500
Salário: 100
Salário: 4500
Salário: 0
 
Salário    - Abono     
R$ 1000.00 - R$  200.00
R$  300.00 - R$  100.00
R$  500.00 - R$  100.00
R$  100.00 - R$  100.00

Foram processados 5 colaboradores
Total gasto com abonos: R$ 1400.00
Valor mínimo pago a 3 colaboradores
Maior valor de abono pago: R$ 900.00

````python
lista_salarios = [1000, 300, 500, 200, 1500, 3000, 3400, 5000, 7000, 2000, 600, 800, 250, 1500, 20000]
abonos = []

print('Salario - Abono')
cont_abono_minimo = 0

for salario in lista_salarios:
    # cada funcionario recebe equivalente a 20% do seu salario bruto de dezembro
    abono_salarial = salario * 0.2
    # o piso do abono será de 100 reais -> funcionarios com salario muito baixo, recebem este valor de piso
    if abono_salarial < 100:
        abono_salarial = 100
    abonos.append(abono_salarial)
    if abono_salarial == 100:
        cont_abono_minimo += 1
    print(f'R$ {salario} - R$ {abono_salarial:,}')

# quantidade de colaboradores
qtde_salarios = len(lista_salarios)
# total gasto com abono
total_gasto_abono = sum(abonos)
# maior valor de abono
maior_valor_abono_pago = max(abonos)

print(f'\nForam processados {qtde_salarios} colaboradores.')
print(f'Total gasto com abonos: R$ {total_gasto_abono:,}')
print(f'Valor mínimo pago a {cont_abono_minimo} colaboradores.')
print(f'Maior valor de abono pago: R$ {maior_valor_abono_pago:,}.')
    
````

#### 7. Uma empresa paga seus vendedores com base em comissões. O vendedor recebe \\$200 por semana mais 9 por cento de suas vendas brutas daquela semana. Por exemplo, um vendedor que teve vendas brutas de \\$3000 em uma semana recebe \\$200 mais 9 por cento de \\$3000, ou seja, um total de \\$470. Escreva um programa (usando um array de contadores) que determine quantos vendedores receberam salários nos seguintes intervalos de valores:
<pre>
$200 - $299
$300 - $399
$400 - $499
$500 - $599
$600 - $699
$700 - $799
$800 - $899
$900 - $999
$1000 em diante
</pre>

Existem várias formas de fazer. Faça primeiro da forma que parecer mais intuitiva para você.

Em seguida, caso queira um desafio:
Desafio: Crie uma forma para chegar na posição da lista a partir do salário, sem fazer vários ifs aninhados.

````python
# lista de vendas semanais de cada vendedor
vendas = [1000, 2000, 3000, 4000, 5000, 6000, 5500, 4500, 3600]

# Definindo os intervalos de salários
intervalos = [
    (200, 299),
    (300, 399),
    (400, 499),
    (500, 599),
    (600, 699),
    (700, 799),
    (800, 899),
    (900, 999),
    # intervalo criado para valor de 1000 em diante
    (1000, float('inf'))
]

# usando uma lista, que aqui chamamos de contador_intervalos para armazenar o número de vendedores em cada intervalo.
contador_intervalos = [0] * len(intervalos)

# nesta parte: acessar cada valor de venda, calcular o salário e em seguida atualiza o intervalo correspondente 
# ao salario do vendedor
for venda in vendas:
    salario_vendedor = 200 + (0.09 * venda)
    for i, (inicio, fim) in enumerate(intervalos):
        if inicio <= salario_vendedor <= fim:
            contador_intervalos[i] += 1
            break

# exibindo resultados
for i, (inicio, fim) in enumerate(intervalos):
    if fim == float('inf'):
        print(f'{inicio} em diante: {contador_intervalos[i]}')
    else:
        print(f'{inicio} - {fim}: {contador_intervalos[i]}')
````
