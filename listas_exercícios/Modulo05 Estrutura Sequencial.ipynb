#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
graus_farenheit = float(input('Insira a temperatura em graus Farenheit: '))

conversor_farenheit_celsius = (graus_farenheit - 32) * (5/9)

print(f'A temperatura em graus Celsius é de {conversor_farenheit_celsius:.2f} °C')
```

#### 9. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
$F = \frac{9}{5}C + 32$

```python
graus_celsius = float(input('Insira a temperatura em graus Celsius: '))

conversor_celsius_farenheit = (9/5) * graus_celsius + 32

print(f'A temperatura em graus Farenheit é de {conversor_celsius_farenheit:.2f} °F')
```

#### 11. Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
##### a. Para homens: $P = 72,7h - 58$
##### b. Para mulheres: $P = 62,1h - 44,7$

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
ganho_hora = float(input('Insira o quanto você ganha por hora trabalhada: '))
numero_horas_trabalhadas_mes = float(input('Insira a quantidade de horas trabalhadas no mês: '))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_bruto = numero_horas_trabalhadas_mes * ganho_hora
print(f'O salário bruto é de R$ {salario_bruto}')
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
desconto_ir = 0.11 * salario_bruto
print(f'O desconto do IR é de R$ {desconto_ir:.2f}')
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
desconto_inss = 0.08 * salario_bruto
print(f'O desconto do INSS é de R$ {desconto_inss:.2f}')
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
desconto_sindicato = 0.05 * salario_bruto
print(f'O desconto do sindicato é de R$ {desconto_sindicato}')
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
descontos = desconto_sindicato + desconto_inss + desconto_ir
salario_liquido = salario_bruto - descontos

print(f'O salário líquido é de R$ {salario_liquido:.2f}')
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
area = float(input('Insira o tamanho da área a ser pintada em m²: '))

# resultado da regra de três
litros_tinta = area / 3

latas = litros_tinta / 18

preco_latas = latas * 80

print(f'Serão necessárias {latas:.2f} latas, resultando em R$ {preco_latas:.2f} no total')
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tam_arquivo = float(input('Insira o tamanho do arquivo em MB: '))
velocidade_internet = float(input('Insira a velocidade da internet em Mbps: '))

# transformando MB para Mbps, para que o tamanho e a velocidade estejam na mesma unidade
tamanho_arq_megabit = tam_arquivo * 8

tempo_download = tamanho_arq_megabit / velocidade_internet

# passando o tempo para minutos
tempo_download_minutos = tempo_download / 60

print(f'O tempo do download será de {tempo_download_minutos:.2f} minutos')
```
