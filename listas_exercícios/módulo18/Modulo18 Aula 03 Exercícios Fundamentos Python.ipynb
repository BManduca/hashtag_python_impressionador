### 1- Imposto a pagar no Lucro Presumido

- 5% sobre faturamento de ISS (mensal)
- 0,65% de PIS sobre faturamento, (mensal)
- 3% de COFINS sobre faturmaneto, (mensal)
- 4.8% de IR (trimestral)
- 10% de IR Adicional sobre o que ultrapassar 20mil do faturamento (trimestral)
- CSLL: 2,88% sobre faturamento (trimestral)

```python
import locale

# Configurar para o padrão brasileiro
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

resultado = {}

faturamento = {
    'jan': 'R$ 95.141,98',
    'fev': 'R$ 95.425,16',
    'mar': 'R$ 89.716,31',
    'abr': 'R$ 78.459,99',
    'mai': 'R$ 71.087,28',
    'jun': 'R$ 83.911,06',
    'jul': 'R$ 56.467,26',
    'ago': 'R$ 88.513,58',
    'set': 'R$ 66.552,49',
    'out': 'R$ 80.164,07',
    'nov': 'R$ 66.964,33',
    'dez': 'R$ 71.525,25',
}

# para cada mês com faturamento no dicionario:
    # - transformar o faturamento de texto para valor numerico
    # - calcular o imposto mensal
    # - calcular o imposto trimestral
    # - adicionar os valores de faturamento, imposto mensal, impostro trimestral no dicionário criado

def transformar_texto_numero(texto):
    texto = texto.replace('R$ ', '')
    texto = texto.replace(' ', '')
    texto = texto.replace('.', '')
    texto = texto.replace(',', '.')
    valor_final = float(texto)

    return valor_final

def calcular_imposto_mensal(valor_faturamento):
    valor_iss = 0.05 * valor_faturamento
    valor_pis = 0.0065 * valor_faturamento
    valor_cofins = 0.03 * valor_faturamento

    imposto_mensal = valor_iss + valor_pis + valor_cofins

    return imposto_mensal

def calcular_imposto_trimestral(valor_faturamento):
    valor_ir = valor_faturamento * 0.048
    valor_csll = valor_faturamento * 0.0288
    adicional_ir = 0

    if valor_faturamento > 20000:
        adicional_ir = (valor_faturamento - 20000) * 0.1

    imposto_trimestral = valor_ir + valor_csll + adicional_ir

    return imposto_trimestral


for mes in faturamento:
    valor_faturamento_mes = transformar_texto_numero(faturamento[mes])
    valor_imposto_mensal_mes = calcular_imposto_mensal(valor_faturamento_mes)
    valor_imposto_trimestral_mes = calcular_imposto_trimestral(valor_faturamento_mes)

    resultado[mes] = (valor_faturamento_mes, valor_imposto_mensal_mes, valor_imposto_trimestral_mes)

# você precisa inserir no sistema um dicionário no formato:

# resultado = {
#     'mes': (faturamento, imposto_mensal, imposto_trimestral),
# }

print(f'{'Mês':<12} {'Faturamento(R$)':>18} {'Imposto Mensal(R$)':>26} {'Imposto Trimestral(R$)':>29}')
print('='*90)
for mes, (valor_faturamento, imposto_mensal, imposto_trimestral) in resultado.items():
    print(f'{mes:<12} {valor_faturamento:>12,.2f} {imposto_mensal:>22,.2f} {imposto_trimestral:>26,.2f}')
# print(resultado)
```
