```python
# importações
import time
import pyautogui

# CONFIGURAÇÕES INICIAIS
pyautogui.PAUSE = 2 # tempo de 2s entre cada ação
# imagens para pesquisa e acesso ao gmail
IMAGENS = [
    'busca_gmail.png',
    'busca_gmail2.png'
]

# ------------------- FUNÇÕES UTILITÁRIAS -------------------

def log(msg, tipo="info"):
    emoji = {"info": "ℹ️", "success": "✅", "error": "❌", "warn": "⚠️", "wait": "⏳"}.get(tipo, "")
    print(f"[{tipo.upper()}] {emoji} {msg}")


def abrir_navegador():
    ''' ACESSAR O NAVEGADOR ARC USANDO RAYCAST '''
    pyautogui.hotkey('command', 'space')
    time.sleep(1)
    pyautogui.write('Arc')
    pyautogui.press('return')
    # Abrindo uma nova aba
    pyautogui.hotkey('command', 't')


def pesquisar_gmail():
    ''' pesquisar pelo Gmail no navegador '''
    pyautogui.write('acessar gmail')
    pyautogui.press('return')
    time.sleep(2)

def clicar_centro_imagem(location):
    ''' Efetuando clique centralizado na localização passada por parametro '''
    # calculando o centro da imagem encontrada
    x, y, width, height = location
    center_x = x + width // 2
    center_y = y + height // 2

    print(f"Coordenadas: x={center_x}, y={center_y}\n")
    
    # movendo o mouse gradualmente para a posição
    pyautogui.moveTo(center_x, center_y, duration=0.5)
    time.sleep(0.5)

    # clicando no link
    pyautogui.click()


def encontrar_e_clicar_imagem_gmail():
    ''' LOCALIZA E CLICA NO LINK DA IMAGEM DO GMAIL '''

    log("Procurando pelo link do Gmail...", "info")
    # print('Procurando pelo link do Gmail...')
    time.sleep(1.5)

    location = None

    # varrendo minha lista de imagens e encontrando a localizacao da imagem
    for img in IMAGENS:
        print(f'Tentando localizar: {img}')
        location = pyautogui.locateOnScreen(img, grayscale=True, confidence=0.7)
        if location:
            log(f"\nImagem {img} encontrada com sucesso!", "success")
            # print(f"\nImagem {img} encontrada com sucesso! ✅")
            clicar_centro_imagem(location)
            return True
        else:
            log(f"\nImagem {img} não foi encontrada!", "warn")
            # print(f"Imagem {img} não foi encontrada! ⚠️")
            
    return False


def encontrar_e_clicar_imagem(imagem):
    location = None

    try:
        location = pyautogui.locateOnScreen(imagem, grayscale=False, confidence=0.8)
        if location:
            log(f"\nImagem {imagem} encontrada com sucesso!", "success")
            # print(f"Imagem {imagem} encontrada com sucesso! ✅")
            clicar_centro_imagem(location)
            return True
            
    except pyautogui.ImageNotFoundException:
        log(f"\nImagem {imagem} não foi encontrada!", "warn")
        # print(f"Imagem {imagem} não foi encontrada! ⚠️")
            
    return False

def wait_load_page(imagem, timeout=15):
    start = time.time()
    while time.time() - start < timeout:
        if pyautogui.locateOnScreen(imagem, grayscale=True, confidence=0.6):
            log("Página carregada com sucesso!", "success")
            # print("Página carregada com sucesso! ✅\n")
            return True
        time.sleep(2)
    log("Tempo de espera esgotado!", "error")
    # print("Tempo de esperar esgotado! ❌")
    return False



if __name__ == "__main__":
    abrir_navegador()
    pesquisar_gmail()
    encontrar_e_clicar_imagem_gmail()
    wait_load_page('logo_gmail_and_search.png')
    encontrar_e_clicar_imagem('menu_google_apps.png',)
    time.sleep(2)
    encontrar_e_clicar_imagem('contatos.png')
    wait_load_page('carregar_tela_contatos.png')
    encontrar_e_clicar_imagem('exportar.png')
    time.sleep(1)
    encontrar_e_clicar_imagem('selecionar_opcao_contato.png')
    time.sleep(1)
    encontrar_e_clicar_imagem('selecionar_lista.png')
    time.sleep(1)
    encontrar_e_clicar_imagem('confirmar_exportar.png')
```

```python
# importações
import time
import pyautogui

# CONFIGURAÇÕES INICIAIS
pyautogui.PAUSE = 2 # tempo de 2s entre cada ação
# imagens para pesquisa e acesso ao gmail
IMAGENS = [
    'busca_gmail.png',
    'busca_gmail2.png'
]

# ------------------- FUNÇÕES UTILITÁRIAS -------------------

def log(msg, tipo="info"):
    emoji = {"info": "ℹ️", "success": "✅", "error": "❌", "warn": "⚠️", "wait": "⏳"}.get(tipo, "")
    print(f"[{tipo.upper()}] {emoji} {msg}")


def abrir_navegador():
    ''' ACESSAR O NAVEGADOR ARC USANDO RAYCAST '''
    log("Abrindo navegador Arc via Raycast...", "info")
    pyautogui.hotkey('command', 'space')
    time.sleep(1)
    pyautogui.write('Arc')
    pyautogui.press('return')
    # Abrindo uma nova aba
    pyautogui.hotkey('command', 't')


def pesquisar_gmail():
    ''' pesquisar pelo Gmail no navegador '''
    log("Pesquisando Gmail no navegador...", "info")
    pyautogui.write('acessar gmail')
    pyautogui.press('return')
    time.sleep(2)


def clicar_centro_imagem(location):
    ''' Efetuando clique centralizado na localização passada por parametro '''
    # calculando o centro da imagem encontrada
    x, y, width, height = location
    center_x = x + width // 2
    center_y = y + height // 2

    log(f"Coordenadas: x={center_x}, y={center_y}\n", "info")
    # print(f"Coordenadas: x={center_x}, y={center_y}\n")
    
    # movendo o mouse gradualmente para a posição
    pyautogui.moveTo(center_x, center_y, duration=0.5)
    time.sleep(0.5)

    # clicando no link
    pyautogui.click()


def encontrar_e_clicar_imagem_gmail():
    ''' LOCALIZA E CLICA NO LINK DA IMAGEM DO GMAIL '''

    log("Procurando pelo link do Gmail...", "info")
    # print('Procurando pelo link do Gmail...')
    time.sleep(1.5)

    location = None

    # varrendo minha lista de imagens e encontrando a localizacao da imagem
    for img in IMAGENS:
        print(f'Tentando localizar: {img}')
        location = pyautogui.locateOnScreen(img, grayscale=True, confidence=0.7)
        if location:
            log(f"\nImagem {img} encontrada com sucesso!", "success")
            # print(f"\nImagem {img} encontrada com sucesso! ✅")
            clicar_centro_imagem(location)
            return True
        else:
            log(f"\nImagem {img} não foi encontrada!", "warn")
            # print(f"Imagem {img} não foi encontrada! ⚠️")
            
    return False


def encontrar_e_clicar_imagem(imagem, mensagem=None, tempo_espera=1, confidence=0.8):

    if mensagem:
        log(mensagem, "info")

    location = None

    try:
        location = pyautogui.locateOnScreen(imagem, grayscale=False, confidence=confidence)
        if location:
            log(f"\nImagem {imagem} encontrada com sucesso!", "success")
            # print(f"Imagem {imagem} encontrada com sucesso! ✅")
            clicar_centro_imagem(location)
            time.sleep(tempo_espera)
            return True
        
    except pyautogui.ImageNotFoundException:
        log(f"\nImagem {imagem} não foi encontrada!", "error")
        # print(f"Imagem {imagem} não foi encontrada! ⚠️")
            
    return False

def wait_load_page(imagem, timeout=15):
    log(f"Aguardando carregamento da página com a imagem '{imagem}'...", "info")
    start = time.time()
    while time.time() - start < timeout:
        if pyautogui.locateOnScreen(imagem, grayscale=True, confidence=0.6):
            log("Página carregada com sucesso!", "success")
            # print("Página carregada com sucesso! ✅\n")
            return True
        time.sleep(2)
    log("Tempo de espera esgotado! A Página não carregou...", "error")
    # print("Tempo de esperar esgotado! ❌")
    return False



if __name__ == "__main__":
    abrir_navegador()
    pesquisar_gmail()
    encontrar_e_clicar_imagem_gmail()
    wait_load_page('logo_gmail_and_search.png')
    encontrar_e_clicar_imagem('menu_google_apps.png',)
    time.sleep(2)
    encontrar_e_clicar_imagem('contatos.png')
    wait_load_page('carregar_tela_contatos.png')
    encontrar_e_clicar_imagem('exportar.png')
    time.sleep(1)
    encontrar_e_clicar_imagem('selecionar_opcao_contato.png')
    time.sleep(1)
    encontrar_e_clicar_imagem('selecionar_lista.png')
    time.sleep(1)
    encontrar_e_clicar_imagem('confirmar_exportar.png')
```
