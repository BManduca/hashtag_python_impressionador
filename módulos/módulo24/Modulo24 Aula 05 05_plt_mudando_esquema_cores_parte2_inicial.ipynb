## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
!pip3 install matplotlib
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20] # milhões de reais
vendas_loja = [10, 8, 6, 4] # milhões de reais
```

```python
plt.bar(anos, vendas_online)

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

plt.xticks(ticks=anos, labels=anos)
plt.ylabel('Vendas (em milhões de reais)')
plt.title('Vendas Online vs. Vendas Lojas')

plt.legend()

plt.show()
```

```python
cores = plt.get_cmap('Pastel2').colors

plt.bar(anos, vendas_online, label='Online', color=cores[0])
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja', color=cores[1])

plt.xticks(ticks=anos, labels=anos)
plt.ylabel('Vendas (em milhões de reais)')
plt.title('Vendas Online vs. Vendas Lojas')

plt.legend()

plt.show()
```

```python
from cycler import cycler
# cycler da uma idéia de ciclos, aonde o matplotlib tem uma função que permite passar por um ciclo de cores
# por exemplo ao passar uma paleta de cores, o mesmo vai passar por ela formando um ciclo e repetetindo enquanto for necessário,
# para construir os elementos do gráfico

cores_cycler = cycler('color', plt.get_cmap('Pastel2').colors)
# rc -> run time configuration -> configuração em tempo de execução
plt.rc('axes', prop_cycle=cores_cycler)

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

plt.xticks(ticks=anos, labels=anos)
plt.ylabel('Vendas (em milhões de reais)')
plt.title('Vendas Online vs. Vendas Lojas')

plt.legend()

plt.show()
```
