```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_xlabel('Mês')
ax.set_ylabel('Vendas (em milhares de reais)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```

## Gráfico de barras

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

fig, ax = plt.subplots()

ax.bar(produtos, quantidades)
ax.set_xlabel('Produto')
ax.set_ylabel('Quantidade de vendas')
ax.set_title('Quantidade de vendas por produto')

plt.show()
```

## Gráfico de pizza

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades_vendidas = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

fig, ax = plt.subplots()

ax.pie(quantidades_vendidas, labels=categorias, explode=destaque, autopct='%.1f%%')
ax.set_title('Categorias de produtos vendidos')

plt.show()
```

## Gráfico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label='Online')
ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

ax.set_xlabel('Ano')
ax.set_ylabel('Vendas (em milhões de reais)')
ax.set_title('Vendas online vs loja')

ax.set_xticks(anos)

ax.legend()

plt.show()
```

## Mais de um sistema de eixos em uma figura

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# gráfico 1 - linha
axs[0, 0].plot(meses, vendas, linewidth=3)
axs[0, 0].tick_params(axis='x', length=0)
axs[0, 0].set_ylabel('Vendas (em milhares de reais)')
axs[0, 0].set_title('Crescimento das Vendas ao Longo do Ano', fontweight='bold')

# gráfico 2 - barras

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

axs[0, 1].bar(produtos, quantidades)
axs[0, 1].tick_params(axis='x', length=0)
axs[0, 1].set_ylabel('Quantidade de vendas')
axs[0, 1].set_title('Quantidade de vendas por produto', fontsize=14, fontweight='bold')

# gráfico 3 - pizza

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades_vendidas = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

axs[1, 0].pie(quantidades_vendidas, labels=categorias, explode=destaque, autopct='%.1f%%')
axs[1, 0].set_title('Categorias de produtos vendidos')

# gráfico 4 - barras empilhadas
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

barras_online = axs[1, 1].bar(anos, vendas_online, label='Online')
barras_loja = axs[1, 1].bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

axs[1, 1].bar_label(barras_online, labels=vendas_online, label_type='center', fontweight='bold')
axs[1, 1].bar_label(barras_loja, labels=vendas_loja, label_type='center', fontweight='bold')

axs[1, 1].set_ylabel('Vendas (em milhões de reais)')
axs[1, 1].set_title('Vendas online vs loja', fontsize=14, fontweight='bold')

axs[1, 1].set_xticks(anos)
axs[1, 1].tick_params(axis='x', length=0)

axs[1, 1].legend()

# ajustes finais
fig.suptitle('Análise de vendas ao longo do ano', fontsize=16, fontweight='bold')

plt.show()
```
