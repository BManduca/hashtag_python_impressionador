```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas, marker='o', markersize=10, linestyle='--', color='green', linewidth='2')

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')

plt.title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas)

plt.annotate('Menor valor', color='red', xy=(2.1, 78), xytext=(3, 80), arrowprops=dict(facecolor='red'))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')

plt.title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# mesmo o vendas sendo uma lista, mas o where vai acabar convertendo para um array
# pois, o numpy trabalha com arrays
marcadores = np.where(vendas == np.min(vendas), 'o', '')

plt.plot(meses, vendas)

# enumerate -> associa um valor a cada posicao de uma lista, tupla...
for posicao, marcador in enumerate(marcadores):
    if marcador == 'o':
        cor='red'
        tam = 80
    else:
        cor='blue'
        tam = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tam, color=cor)

# plt.annotate('Menor valor', color='red', xy=(2.2, 78), xytext=(3, 80), arrowprops=dict(facecolor='red'))
plt.annotate('Menor valor', color='red', xy=(2.2, 78), xytext=(2.3,78))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')

plt.title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```
