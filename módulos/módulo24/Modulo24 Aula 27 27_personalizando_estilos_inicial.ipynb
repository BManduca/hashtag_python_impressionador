## Personalizando um estilo pré-pronto

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

plt.style.use('fivethirtyeight')

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
# criando uma lista de tupla chamada dados_ordenados e ordenando a mesma de maneira crescente
dados_ordenados = sorted(zip(quantidades, produtos))
lista_quantidades_ordenados = [x[0] for x in dados_ordenados]
lista_produtos_ordenados = [x[1] for x in dados_ordenados]

plt.style.use('fivethirtyeight')

barras = plt.barh(lista_produtos_ordenados, lista_quantidades_ordenados)

plt.bar_label(barras, labels=lista_quantidades_ordenados, padding=3, fontsize=10, fontweight='bold')

barras[4].set_color('C3')
barras[0].set_color('C1')

# plt.xlabel('Quantidade Vendida')
# plt.ylabel('Produto')
plt.xticks([])
plt.tick_params(axis='y', length=0)
plt.title('Quantidade de Produtos Vendidos', color='dimgray')
plt.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
# criando uma lista de tupla chamada dados_ordenados e ordenando a mesma de maneira crescente
dados_ordenados = sorted(zip(quantidades, produtos))
lista_quantidades_ordenados = [x[0] for x in dados_ordenados]
lista_produtos_ordenados = [x[1] for x in dados_ordenados]

plt.style.use('fivethirtyeight')

params={
    'axes.titlecolor': 'dimgray', # para o sistema de eixos (axes), utilizar a cor XXXX para o title
    'figure.figsize': (12, 6)
}

plt.rcParams.update(params)

barras = plt.barh(lista_produtos_ordenados, lista_quantidades_ordenados)

plt.bar_label(barras, labels=lista_quantidades_ordenados, padding=3, fontsize=10, fontweight='bold')

barras[4].set_color('C3')
barras[0].set_color('C1')

# plt.xlabel('Quantidade Vendida')
# plt.ylabel('Produto')
plt.xticks([])
plt.tick_params(axis='y', length=0)
plt.title('Quantidade de Produtos Vendidos', color='dimgray')
plt.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8, 4))

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

ax.plot(meses, vendas, marker='o')
ax.set_xlabel('Mês')
ax.set_ylabel('Vendas (em milhares)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')
ax.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

params = {
    'axes.titlecolor': 'dimgray',
    'figure.figsize': (12, 6)
}

plt.rcParams.update(params)

fig, ax = plt.subplots()

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

ax.plot(meses, vendas, marker='o')

for posicao in (0, -1):
    ax.scatter(meses[posicao], vendas[posicao], color='red', zorder=2)
    ax.annotate(vendas[posicao], color='dimgray', xy=(meses[posicao], vendas[posicao]), xytext=(-8, 10), textcoords='offset points', size=12, weight='bold')

ax.set_xlabel('Mês')
ax.set_ylabel('Vendas (em milhares)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')
ax.grid(False)

plt.show()
```
