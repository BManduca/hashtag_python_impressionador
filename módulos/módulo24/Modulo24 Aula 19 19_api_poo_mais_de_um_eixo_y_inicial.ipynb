## Mais de um eixo y em uma figura

Crie uma figura de um gráfico de precipitação com um eixo secundário para a temperatura, para mostrar a relação entre a temperatura e a precipitação em uma cidade.



```python
!pip3 install matplotlib
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
import matplotlib.pyplot as plt

cor_temperatura = 'tab:red'
cor_precipitacao = 'tab:blue'


# criação do sistema de eixos
fig, ax1 = plt.subplots()


# gerando grafico de barras com relação a precipitação média
barras_precipitacao = ax1.bar(meses, precipitacao_media, color=cor_precipitacao, label='Precipitação média (mm)')
ax1.set_ylabel('Precipitação média (mm)', color=cor_precipitacao)
ax1.tick_params(axis='x', length=0)
ax1.tick_params(axis='y', labelcolor=cor_precipitacao)
ax1.bar_label(barras_precipitacao, labels=precipitacao_media, label_type='center', fontweight='bold')
# ax1.set_ylim(0, 250)


# gerando gráfico de linha para a temperatura media
# criando uma duplicata ou um gemeo, designando ao ax2
ax2 = ax1.twinx()
ax2.plot(meses, temperatura_media, color=cor_temperatura, linewidth=2, label='Temperatura média (˚C)')
ax2.set_ylabel('Temperatura média (˚C)', color=cor_temperatura)
ax2.tick_params(axis='y', labelcolor=cor_temperatura)
ax2.set_ylim(20, 30)

ax1.set_title('Temperatura e precipitação ao longo do ano', fontsize=15, fontweight='bold')

# bbox_transform=ax1.transAxes -> transformando o sistema de coordenadas e levando
# em consideração o sistema de coordendas do ax1
fig.legend(bbox_to_anchor = (0.75, 1), bbox_transform=ax1.transAxes)

plt.savefig('imagens/temperatura_precipitacao_longo_ano.png', bbox_inches='tight', dpi=300)

plt.show()
```
