```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

# dados
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# usando o np where, crie uma lista de marcadores o para o valor mínimo e vazio para o resto
marcadores = np.where(vendas == np.min(vendas), 'o', '')

# criando a paleta de cores
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

# criando o grafico base, de acordo com os meses e as vendas
plt.plot(meses, vendas)

# definindo marcadores e anotações
for posicao, marcador in enumerate(marcadores):
    if marcador == 'o':
        cor = 'red'
        tamanho = 80
    else:
        cor = 'blue'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)

# adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(2.3, 77.5))

# destacando região
# axvspan |  ax -> eixo x | v -> vertical | span -> a regiao aonde sera feito o destaque
plt.axvspan(6, 11, alpha=0.3, color=cores[1])

plt.text(6.4, 80, 'Recuperação de crescimento', color='green', fontsize=8, weight='bold')

# nomeando os eixos
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')

# título do gráfico
plt.title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

# dados
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# usando o np where, crie uma lista de marcadores o para o valor mínimo e vazio para o resto
marcadores = np.where(vendas == np.min(vendas), 'o', '')

# criando a paleta de cores
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

# criando o grafico base, de acordo com os meses e as vendas
plt.plot(meses, vendas)

# definindo marcadores e anotações
for posicao, marcador in enumerate(marcadores):
    if marcador == 'o':
        cor = 'red'
        tamanho = 90
    else:
        cor = 'darkblue'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-6, 8), textcoords='offset points')

# adicionando uma anotação para mostrar o menor valor em março
# plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(2.3, 77.5))

# destacando região
# axvspan |  ax -> eixo x | v -> vertical | span -> a regiao aonde sera feito o destaque
plt.axvspan(6, 11, alpha=0.3, color=cores[1])

plt.text(6.4, 80, 'Recuperação de crescimento', color='green', fontsize=8, weight='bold')

plt.box(False)
plt.tick_params(axis='x', length=0)
plt.yticks([])

# nomeando os eixos
plt.ylabel('Vendas (em milhares)')

# título do gráfico
plt.title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```
