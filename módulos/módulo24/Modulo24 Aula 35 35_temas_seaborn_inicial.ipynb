# Temas do Seaborn

```python
import seaborn as sns

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

dados = {'Meses': meses, 'Vendas': vendas}

sns.lineplot(x='Meses', y='Vendas', data=dados);
```

No Seaborn, um tema é a combinação de estilo, contexto e paleta de cores que juntos definem a aparência de um gráfico.

https://seaborn.pydata.org/generated/seaborn.set_theme.html

https://seaborn.pydata.org/tutorial/aesthetics.html

https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette

https://seaborn.pydata.org/tutorial/color_palettes.html

```python
estilos = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
contextos = ['paper', 'notebook', 'talk', 'poster']
paletas = ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']  # além das do matplotlib
```

```python
import seaborn as sns

sns.set_theme()

sns.lineplot(x='Meses', y='Vendas', data=dados);
```

```python
import seaborn as sns

'''
paper -> artigo
talk -> palestra/apresentação
poster -> pôster
'''
sns.set_theme(style='darkgrid', context='notebook')

lines = sns.lineplot(x='Meses', y='Vendas', data=dados)
lines.set_xlabel('')
lines.set_title('Vendas ao longo do ano', fontsize=15, fontweight='bold');
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, eixo = plt.subplots(figsize=(12,4))

'''
paper -> artigo
talk -> palestra/apresentação
poster -> pôster
'''
sns.set_theme(style='darkgrid', context='poster')

lines = sns.lineplot(x='Meses', y='Vendas', data=dados, ax=eixo)

lines.set_title('Vendas ao longo do ano', fontsize=16, weight='bold')
lines.set_xlabel('')

plt.show()
```

```python
import seaborn as sns

'''
paper -> artigo
talk -> palestra/apresentação
poster -> pôster
'''
sns.set_theme(style='darkgrid', context='notebook', palette='bright')

lines = sns.lineplot(x='Meses', y='Vendas', data=dados)
lines.set_xlabel('')
lines.set_title('Vendas ao longo do ano', fontsize=15, fontweight='bold');
```

```python
import seaborn as sns

sns.set_theme()

# Dados fictícios
cidades = ['Brasília', 'Rio de Janeiro', 'São Paulo', 'Curitiba', 'Manaus']
temperaturas = [27, 29, 25, 23, 31]  # Temperaturas médias anuais em graus Celsius

dados_temperaturas = {'Cidade': cidades, 'Temperatura': temperaturas}

barras = sns.barplot(x='Cidade', y='Temperatura', data=dados_temperaturas, hue=cidades)

barras.set_ylabel('Temperatura em ˚C')
barras.set_xlabel('')
barras.set_title('Temperaturas médias anuais');
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid', palette='Paired', context='talk')

fig, eixo = plt.subplots(figsize=(10,6))

# Dados fictícios
cidades = ['Brasília', 'Rio de Janeiro', 'São Paulo', 'Curitiba', 'Manaus']
temperaturas = [27, 29, 25, 23, 31]  # Temperaturas médias anuais em graus Celsius

dados_temperaturas = {'Cidade': cidades, 'Temperatura': temperaturas}

sns.barplot(x='Cidade', y='Temperatura', data=dados_temperaturas, hue=cidades)

plt.xticks(rotation=30, ha='right')

eixo.set_ylabel('Temperatura em ˚C')
eixo.set_xlabel('')
eixo.set_title('Temperaturas médias anuais', fontsize=16, fontweight='bold')

plt.show()
```
