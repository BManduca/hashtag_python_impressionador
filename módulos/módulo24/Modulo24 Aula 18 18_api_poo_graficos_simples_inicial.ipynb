```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
# subplots -> criar um sistema de eixos
fig, ax = plt.subplots()

ax.plot(meses, vendas)

# configurando rotulos e títulos
ax.set_xlabel('Mês')
ax.set_ylabel('Vendas (milhares de reais)')
ax.set_title('Evolução de vendas ao longo do ano', fontsize=16, fontweight='bold')

plt.show()
```

## Gráfico de barras

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
fig, ax = plt.subplots()

barras = ax.bar(produtos, quantidades, label='Qtde. de Produtos')

# configurando labels e títulos
ax.set_title('Relação Produtos vs. Quantidades', fontsize=16, fontweight='bold')
ax.bar_label(barras, labels=quantidades, label_type='center', fontweight='bold')

ax.tick_params(axis='x', length=0)
ax.set_yticks([])

plt.box(False)
plt.legend()

# solicitando que o matplotlib apresente tudo que foi configurado anteriormente
plt.show()
```

## Gráfico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]
```

```python
fig, ax = plt.subplots()

barras_online = ax.bar(anos, vendas_online, label='Online')
barras_loja = ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

ax.bar_label(barras_online, labels=vendas_online, label_type='center', fontweight='bold')
ax.bar_label(barras_loja, labels=vendas_loja, label_type='center', fontweight='bold')

# apresentando os anos como números inteiros
ax.set_xticks(anos)

ax.set_yticks([])
ax.tick_params(axis='x', length=0)
ax.set_title('Vendas (em milhões) Online vs. Lojas', fontsize=14, fontweight='bold')
ax.legend()

plt.box(False)


```
