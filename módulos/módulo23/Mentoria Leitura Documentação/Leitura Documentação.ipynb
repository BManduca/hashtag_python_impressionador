```python
# pandas
import pandas as pd

dataframe_passengers_titanic = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)

display(dataframe_passengers_titanic)

# pd.DataFrame -> pd é o alias denominado ao importar a biblioteca do pandas
# pd.DataFrame -> recebe um dicionário como parâmetro e cria a tabela a partir do dicionário
```

```python
# quando selecionamos uma única coluna no pandas, o resultado é um pandas series
display(dataframe_passengers_titanic['Age'])

idade_max = dataframe_passengers_titanic['Age'].max()

print(f'A maior idade entre os passageiros do Titanic é {idade_max} anos')
```

```python
# funcao datetime

from datetime import datetime
import locale
import time

locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')

today = datetime.now()

date_today = datetime.strptime('27/12/2024 22:14', '%d/%m/%Y %H:%M')
date_today_formatado = date_today.strftime('%A, %d de %B de %Y as %H:%M')
print(date_today_formatado)

print(today)
print(date_today)


tempo_em_struct = time.localtime()
print(tempo_em_struct)
# formatando
tempo_formatado = time.strftime('%A, %d de %B de %Y, %H:%M', tempo_em_struct)
print(tempo_formatado)

```
