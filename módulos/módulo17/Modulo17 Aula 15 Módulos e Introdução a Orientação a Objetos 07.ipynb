### Outros tipos de Gráfico:

- Linha

```python
# randint -> menor valor, maior valor e qtde 
vendas = np.random.randint(1000, 3000, 50)
# range de números(meses) -> excluindo o último
meses = np.arange(1, 51)
```

```python
plt.plot(meses, vendas)
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses, vendas, color='red')
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.plot()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Barras

```python
plt.bar(meses, vendas)
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
# figsize(largura, altura)
plt.figure(1, figsize=(15, 3))
# subplot(num de linhas, qtde de graficos, posicao do grafico)
plt.subplot(1, 3, 1)
plt.plot(meses, vendas, color='red')
plt.subplot(1, 3, 2)
plt.scatter(meses, vendas)
plt.subplot(1, 3, 3)
plt.bar(meses, vendas)
plt.show()
```
