## Importando e apresentando a base de dados

```python
import pandas as pd

dataset_top_rated_movies = pd.read_csv('top_rated_movies.csv')

display(dataset_top_rated_movies)
```

```python
display(dataset_top_rated_movies.info())
```

## Conversão da coluna release_date do formato aaaa-mm-dd para o formato dd/mm/aaaa

```python
# convertendo a coluna release_date para o formato datetime do pandas
dataset_top_rated_movies['release_date'] = pd.to_datetime(dataset_top_rated_movies['release_date'], format='%Y-%m-%d')

# Formatar a coluna release_date para o format dd/mm/aaaa
dataset_top_rated_movies['release_date'] = dataset_top_rated_movies['release_date'].dt.strftime('%d/%m/%Y')
```

```python
display(dataset_top_rated_movies.info())
```

## Criando nova coluna Year no dataframe

```python
dataset_top_rated_movies['year'] = pd.to_datetime(dataset_top_rated_movies['release_date'], format='%d/%m/%Y').dt.year

dataset_top_rated_movies['year'].astype('int64')

display(dataset_top_rated_movies.info())
```

## Ordenando os filme de acordo com ano

```python
dataset_top_rated_movies_year = dataset_top_rated_movies.sort_values(by='year', ascending=False)

display(dataset_top_rated_movies_year)
```

## Filmes de 2023, com maior popularidade

```python
dataset_top_rated_movies_2023 = dataset_top_rated_movies.loc[dataset_top_rated_movies['year'] == 2023, :]

dataset_top_rated_movies_2023_popularity = dataset_top_rated_movies_2023.sort_values(by='popularity', ascending=False)

display(dataset_top_rated_movies_2023_popularity.head())
```

## Filmes em 2023, com maiores médias e contagem de votos

```python
dataset_top_rated_movies_2023_vote_average = dataset_top_rated_movies_2023.sort_values(by='vote_average', ascending=False)

dataset_top_rated_movies_2023_vote_count = dataset_top_rated_movies_2023_vote_average.sort_values(by='vote_count', ascending=False)

display(dataset_top_rated_movies_2023_vote_count)
```

```python
dataset_movies_popularity = dataset_top_rated_movies_2023[['title','popularity']]

dataset_movies_popularity_head = dataset_movies_popularity.head()

display(dataset_movies_popularity_head)

dataset_movies_popularity_head.plot(kind='bar', title='Os 5 filmes com maiores popularidades')
```
