## Importando e apresentando a base de dados

```python
import pandas as pd

table_car_information = pd.read_csv('sport_car_price.csv')

display(table_car_information)
```

## Informações gerais
- Verificando quantas linhas tem a tabela de maneira geral
- Existência de valores vazios e que necessitem ser tratados

```python
display(table_car_information.info())

print(table_car_information.isnull().sum())
```

## Removendo registros com valores vazios/nulos

```python
table_car_information_clean = table_car_information.dropna()

# apresentando o tamanho do dataset antes e depois de remover registros nulos
print(f'Tamanho antes de retirar registros nulos do dataset: {table_car_information.shape}')
print(f'Tamanho depois de retirar registros nulos do dataset: {table_car_information_clean.shape}')

display(table_car_information_clean.head())
```

## qual carro da marca porsche tem maior Horsepower?

```python
table_porsche = table_car_information_clean.loc[table_car_information_clean['Car Make'] == 'Porsche', :]
table_porsche = table_porsche.loc[table_porsche['Year'] == 2022, :]


table_porsche_high_horsepower = table_porsche[['Car Model', 'Year', 'Horsepower', 'Torque (lb-ft)']].sort_values(by='Horsepower', ascending=False)

display(table_porsche_high_horsepower)
```
